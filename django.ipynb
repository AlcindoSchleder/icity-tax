```python
for col in keys:
    for row in keys:
        tax = float(df[col][row])
        insTax = Taxes.objects.add(
            fk_typetaxes=1,
            fk_countries_origin=55,
            fk_countries_destine=55,
            fk_provinces_origin=col,
            fk_provinces_destine=row,
            taxdef=tax
        )
        ncmcodes = NcmCodes.objects.all().orderby('pk_ncmcodes')
        for ncmcode in ncmcodes:
            Taxes.objects.add(
                fk_taxes=insTax.pk_taxes,
                fk_ncmcodes=ncmcode.pk_ncmcodes,
                tax=tax
            )
```
